```python
import pandas as pd
import numpy as np
import chart_studio.plotly as py
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
```

```python
stock_data = pd.read_csv('OutputData/2019-20_BTC-USD.csv')
stock_data.head(5)
```

```python
%%time
Figure1 = go.Figure()

Figure1.add_trace(go.Scatter(x=stock_data['Date'], y=stock_data['Adj Close'],mode='lines',name='Adj Close'))
Figure1.add_trace(go.Scatter(x=stock_data['Date'], y=stock_data['High'],mode='lines',name='High'))
Figure1.add_trace(go.Scatter(x=stock_data['Date'], y=stock_data['Low'],mode='lines',name='Low'))
Figure1.add_trace(go.Scatter(x=stock_data['Date'], y=stock_data['Open'],mode='lines',name='Open'))
Figure1.add_trace(go.Scatter(x=stock_data['Date'], y=stock_data['Close'],mode='lines',name='Close'))


Figure1.update_layout(
    title="Cryptocurrency market 2019/ January 2020 ",
    xaxis_title="Dates",
    yaxis_title="Price $")
                         
Figure1.show()
```
